# Q.5 - Determine a suitable way to combine the orders_products_combined dataframe with your products data set. 

```python
#Import Products Checked as df_prods
df_prods = pd.read_csv(os.path.join(path, '02 Data', 'Prepared Data', 'products_checked.csv'), index_col=False)
```

### The best way to combine these dfs is to merge them using the product_id column. Column names are identical, so there's no need to rename either.

```python
# Removing columns from merged df
df_op_combo = df_ords_prods_combo.drop(columns = ['Unnamed: 0', 'Unnamed: 0.1', "_merge"])
```

```python
df_op_combo.shape
```

```python
# Removing unnames column from prods df
df_prods_B = df_prods.drop(columns = ['Unnamed: 0'])
```

```python
df_prods_B.shape
```

```python
# Merging dfs with default inner join
df_merged = df_op_combo.merge(df_prods_B, on = 'product_id', indicator = True)
```

```python
# Checking merged value counts
df_merged['_merge'].value_counts()
```

```python
df_merged.head()
```

### Merged df is reduced by 29,630 rows

```python
# Performing outer merge for comparison
df_outer_merged = df_op_combo.merge(df_prods_B, on = 'product_id', indicator = True, how = 'outer')
```

# Q.6 - Confirm the results of the merge using the merge flag.

```python
# Checking merged value counts after outer merge
df_outer_merged['_merge'].value_counts()
```

```python
df_outer_merged.head()
```

```python
# Pulling only the right_only data for a quick look
df_outer_merged[df_outer_merged['_merge'] == 'right_only']
```

```python
# Performing left merge for comparison as well. Outer isn't the best option, but a left merge might work better
df_left_merged = df_op_combo.merge(df_prods_B, on = 'product_id', indicator = True, how = 'left')
```
