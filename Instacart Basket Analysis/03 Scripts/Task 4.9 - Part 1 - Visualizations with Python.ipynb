# Q.5 - Complete fundamental data quality and consistency checks.

```python
# Checking for mixed type columns

for col in cust_df.columns.tolist():
    weird = (cust_df[[col]].applymap(type) != cust_df[[col]].iloc[0].apply(type)).any(axis = 1)
    if len (cust_df[weird]) > 0:
        print (col)
```

```python
# Assigning first_name column as string variable

cust_df['first_name'] = cust_df['first_name'].astype(str)
```

```python
# Checking for null values

cust_df.isnull().sum()
```

```python
# Checking for duplicates

cust_dups = cust_df[cust_df.duplicated()]
```

```python
cust_dups.shape
```

```python
cust_dups.head(10)
```
