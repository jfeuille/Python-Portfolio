```python
# Creating a subset of df_op_mrg
df = df_op_mrg[:1000000]
```

```python
# Defining Price Label column with an if-else statement

def price_label(row):
    
    if row['prices'] <= 5:
        return 'Low-range product'
    elif (row['prices'] > 5) and (row['prices'] <= 15):
        return 'Mid-range product'
    elif row['prices'] > 15:
        return 'High range'
    else: return 'Not enough data'
```

```python
df.head()
```

```python
# Applying function to df subset

df['price_range'] = df.apply(price_label, axis=1)
```

# Q.2 - Suppose your clients have changed their minds about the labels you created in your “busiest_day” column. Now, they want “Busiest day” to become “Busiest days” (plural). This label should correspond with the two busiest days of the week as opposed to the single busiest day. At the same time, they’d also like to know the two slowest days. Create a new column for this using a suitable method.

```python
df_op_mrg['orders_day_of_week'].value_counts(dropna = False)
```

```python
busy_days = []

for value in df_op_mrg['orders_day_of_week']:
    if value == 0:
        busy_days.append('Busiest days')
    elif value == 1:
        busy_days.append('Busiest days')
    elif value == 3:
        busy_days.append('Slowest days')
    elif value == 4:
        busy_days.append('Slowest days')
    elif value == 'NaN':
        busy_days.append('Not applicable')
    else:
        busy_days.append('Regularly busy')
```

```python
busy_days
```

```python
df_op_mrg['busiest_days'] = busy_days
```

```python
df_op_mrg['busiest_days'].value_counts(dropna = False)
```

# Q.4 - Identify the busiest hours of the day. Rather than by hour, they want periods of time labeled “Most orders,” “Average orders,” and “Fewest orders.” Create a new column containing these labels called “busiest_period_of_day.”

```python
df_op_mrg['order_time_of_day'].value_counts(dropna = False)
```

```python
df_op_mrg.shape
```

```python
busy_period = []

for value in df_op_mrg['order_time_of_day']:
    if value in range (10,16):
        busy_period.append('Most orders')
    elif value in range (1,5):
        busy_period.append('Least orders')
    elif value == 'NaN':
        busy_period.append('Not applicable')
    else:
        busy_period.append('Average orders')
```

```python
busy_period
```

```python
df_op_mrg['busiest_period_of_day'] = busy_period
```
